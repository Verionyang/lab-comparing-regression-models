```python
def num_cat_split(df):
    df_num = df.select_dtypes(include = np.number)
    df_cat = df.select_dtypes(include = 'object')
    return df_num, df_cat
```

The pair of monthly premiums of the auto and the total claim has the highest correlation. It suggests that the higher the premium cost is, the higher the total claim amount is. A higher premium implies a better insurance package so the drivers can declare more claims and get more money in return.

```python
def con_dis_split(df, thr=36):
    con = []
    dis = []
    for i in range(df.shape[1]):
        if df.iloc[:,i].unique().size >=thr: 
            con.append(i)
        else: 
            dis.append(i)
    return df.iloc[:,con],df.iloc[:, dis]
```

```python
customer_con,customer_dis = con_dis_split(customer_num)

display(customer_con.head(5))
customer_dis.head(5)
```

```python
def barplot_df(df):
    fig,ax = plt.subplots(1,df.shape[1],figsize=(16,16))
    for i in range(0,df.shape[1]):
        unique_values = df.iloc[:,i].unique().size
        if (unique_values <= 6):
            sns.countplot(x=df.iloc[:,i], order=df.iloc[:,i].value_counts().index,ax=ax[i])
        else:
            sns.countplot(y=df.iloc[:,i], order=df.iloc[:,i].value_counts().index,ax=ax[i])
    return fig
```

```python
fig1 = barplot_df(customer_dis)
```

```python
def histplot_df(df):
    fig,ax = plt.subplots(1,df.shape[1],figsize=(16,16))
    for i in range(0,df.shape[1]):
        sns.histplot(x=df.iloc[:,i],ax=ax[i])
    return fig
```

```python
fig2 = histplot_df(customer_con)
```

I will apply power transformation to the continous variables to change the pattern so they are closer to Gaussian.. 

```python
def boxplot_df(df):
    fig,ax = plt.subplots(1,df.shape[1],figsize=(16,16))
    for i in range(0,df.shape[1]):
        sns.boxplot(y=df.iloc[:,i],ax=ax[i])
    return fig
```

```python
fig3 = boxplot_df(customer_con)
```

```python
def clean_outliers(df,col):
    max_thr = (int(df[col].quantile([0.75]))-int(df[col].quantile([0.25])))*1.5+int(df[col].quantile([0.75]))
    df[col]=df[col][df[col]<=max_thr]
    return df[col]
```

```python
X = data1.drop(columns='total_claim_amount',axis=1)
y = data1['total_claim_amount']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=31)
```

```python
X_train_num = X_train.select_dtypes(include = np.number)
X_test_num = X_test.select_dtypes(include = np.number)
X_train_cat = X_train.select_dtypes(include = 'object')
X_test_cat = X_test.select_dtypes(include = 'object')
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first')
encoder.fit(X_train_cat)
encoded = encoder.transform(X_train_cat).toarray()
X_train_coded = pd.DataFrame(encoded,columns=encoder.get_feature_names_out(),index=X_train_cat.index)
X_train_coded.head()
```

```python
encoded2 = encoder.transform(X_test_cat).toarray()
X_test_coded = pd.DataFrame(encoded2,columns=encoder.get_feature_names_out(),index=X_test_cat.index)
X_test_coded.head()
```

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer()
pt.fit(X_train_num)
X_train_pt_np = pt.transform(X_train_num)
X_train_pt = pd.DataFrame(X_train_pt_np,columns=X_train_num.columns,index=X_train_num.index)
X_train_pt.head()
```

```python
X_test_pt_np = pt.transform(X_test_num)
X_test_pt = pd.DataFrame(X_test_pt_np,columns=X_test_num.columns,index=X_test_num.index)
X_test_pt.head()
```

```python
X_train_new =pd.concat([X_train_pt,X_train_coded],axis=1)
X_test_new = pd.concat([X_test_pt,X_test_coded],axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train_new)

X_train_normalised = scaler.transform(X_train_new)
X_train_new_scaled_df = pd.DataFrame(X_train_normalised,columns=X_train_new.columns,index=X_train_new.index)
X_test_normalised = scaler.transform(X_test_new)
X_test_new_scaled_df = pd.DataFrame(X_test_normalised,columns=X_test_new.columns,index=X_test_new.index)
```

```python
display(X_train_new_scaled_df.head())
X_test_new_scaled_df.head()
```

```python
from sklearn import linear_model
from sklearn.metrics import r2_score
lm = linear_model.LinearRegression()
lm.fit(X_train_new_scaled_df,y_train)
y_pred_test = lm.predict(X_test_new_scaled_df)
r2_score_test = r2_score(y_test,y_pred_test)

print("The R2 score of the test set is {:.2f}".format(r2_score_test))
```

```python
#from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

def linear_model(X1,X2,y1,y2):
    from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
    from sklearn import linear_model
    lm = linear_model.LinearRegression()
    lm.fit(X1,y1)
    y_pred = lm.predict(X2)
    MAE = mean_absolute_error(y2,y_pred)
    MSE = mean_squared_error(y2,y_pred)
    RMSE = np.sqrt(mean_squared_error(y2,y_pred))
    MAPE = mean_absolute_percentage_error(y2,y_pred)
    R2 = r2_score(y2,y_pred)
    
    dic = {"Error_metric":["MAE","MSE","RMSE","MAPE","R2"],"Value":[round(MAE,2),round(MSE,2),round(RMSE,2),round(MAPE,2),round(R2,2)]}
    value = pd.DataFrame(data = dic)
    return value
```

```python
linear_model(X_train_new_scaled_df,X_test_new_scaled_df,y_train,y_test)
```

```python
def model_training(X1,X2,y1,y2):
    from sklearn.neighbors import KNeighborsRegressor
    from sklearn.metrics import r2_score
    from sklearn import linear_model
    from sklearn.neural_network import MLPRegressor
    
    lm = linear_model.LinearRegression()
    lm.fit(X1,y1)
    y_pred = lm.predict(X2)
    R2 = r2_score(y2,y_pred)
    
    neigh = KNeighborsRegressor(n_neighbors=6)
    neigh.fit(X1,y1)
    y_pred_2=neigh.predict(X2)
    R2_KN = r2_score(y2,y_pred_2)
    
    regr = MLPRegressor(random_state=1, max_iter=10000).fit(X1,y1)
    regr.predict(X2)
    score = regr.score(X2,y2)
    
    value_list = [R2,R2_KN,score]
    df = pd.DataFrame(data = value_list, index=['Linear Regression','KNeighborsRegressor','MLPRegressor'],columns=['R2 score'])
    
    return df
```

```python
model_training(X_train_new_scaled_df,X_test_new_scaled_df,y_train,y_test)
```

### According to the (R2) score, the MLPRegressor model performs the best.
